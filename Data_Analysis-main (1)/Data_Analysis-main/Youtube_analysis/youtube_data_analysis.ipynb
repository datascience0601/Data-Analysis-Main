# YOUTUBE DATA ANALYSIS -2024

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('YOUTUBE CHANNELS DATASET.csv')
df
```

```python
df.info()
```

```python
#Deleting all nan values
df.dropna(inplace=True)
df.reset_index(inplace=True)
df.drop('index',axis=1,inplace=True)
df.info()
```

```python
df.head()
```

```python
#changing the data type of Ranking column
df['Ranking']=df['Ranking'].astype(int)
df.info()
```

```python
#Removing unwanted data and converting Subscribers col into float data type
df['Subscribers']=df['Subscribers'].str.strip('M')
df['Subscribers']=df['Subscribers'].astype(float)
df.info()
```

```python
#Replacing ',' to '' and converting uploads and views col into int data type

df['Uploads']=df['Uploads'].str.replace(',','')
df['Uploads']=df['Uploads'].astype(int)

df['Views']=df['Views'].str.replace(',','')
df['Views']=df['Views'].astype(int)
df.info()
```

```python
df.head()
```

```python
df['Country'].unique()
```

```python
#changing the data present in country column
country_codes = {
    'US': 'United States',
    'IN': 'India',
    'JP': 'Japan',
    'KR': 'South Korea',
    'CA': 'Canada',
    'PT': 'Portugal',
    'AE': 'United Arab Emirates',
    'BR': 'Brazil',
    'AR': 'Argentina',
    'PK': 'Pakistan',
    'MX': 'Mexico',
    'IT': 'Italy',
    'GB': 'United Kingdom',
    'RU': 'Russia',
    'ES': 'Spain',
    'ID': 'Indonesia',
    'CL': 'Chile',
    'PH': 'Philippines',
    'TR': 'Turkey',
    'PR': 'Puerto Rico',
    'SV': 'El Salvador',
    'CO': 'Colombia',
    'NO': 'Norway'
}

df['Country']=df['Country'].map(country_codes)
df.info()
```

```python
df.head()
```

# Data Visualization

```python
def data_label(a):
  for i in a.containers:
    a.bar_label(i,label_type='center',rotation=90)
```

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Username',y='Subscribers',data=df.sort_values(by='Subscribers',ascending=False).iloc[:5,:],color='red')
data_label(a)
plt.title('Subscribers by username',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

**channels such as Mr Beast, T-Series, cocomelon-nursery Rhymes are having highest subscribers.**

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Username',y='Views',data=df.sort_values(by='Views',ascending=False).iloc[:5,:],color='red')
data_label(a)
plt.title('views by username',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

**T-series, cocomelon-nursery Rhymes and set india are having highest views.**

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Username',y='Uploads',data=df.sort_values(by='Uploads',ascending=False).iloc[:5,:],color='red')
plt.title('Uploads by username',fontsize=15)
data_label(a)
plt.xticks(rotation=90)
plt.show()
```

**no.of uploads for ABP news, Aaj tak, ABS-CBN entertainment are more.**

```python
plt.figure(figsize=(15,9))
a=sns.countplot(x='Country',data=df,color='red')
data_label(a)
plt.title('Count of Channels by country',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

**India and USA are having highest channels compared to other countries.**

```python
df.sort_values(by='Views',ascending=False)
```

```python
df1=df.pivot_table(values=['Subscribers','Uploads','Views'],index='Country',aggfunc=['sum'])
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Subscribers'),data=df1.sort_values(('sum', 'Subscribers'),ascending=False).iloc[:5,:],color='red')
data_label(a)
plt.title('Total Subscribers by country',fontsize=15)
plt.ylabel('Total subscribers')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Uploads'),data=df1.sort_values(('sum', 'Uploads'),ascending=False).iloc[:5,:],color='red')
data_label(a)
plt.title('Total Uploads by country',fontsize=15)
plt.ylabel('Total Uploads')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
a=sns.barplot(x='Country',y=('sum', 'Views'),data=df1.sort_values(('sum', 'Views'),ascending=False).iloc[:5,:],color='red')
data_label(a)
plt.title('Total Views by country',fontsize=15)
plt.ylabel('Total Views')
plt.xticks(rotation=90)
plt.show()
```

**Total no.of subscribers, views are more for India and United states.**
